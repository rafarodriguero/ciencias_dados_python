```python
df.query('index in @setor')
```

```python
df.query('index in @setor').loc[:,['mrgebit', 'mrgliq', 'roic', 'roe']]
```

```python
varejo = df.query('index in @setor').loc[:,['mrgebit', 'mrgliq', 'roic', 'roe']]
varejo
```

```python
fig = make_subplots(rows=2, 
                    cols = 2, 
                    row_heights=[5,5],
                    #column_widths=[3,3], 
                    subplot_titles=('Margem Ebtida', 'Margem Liquida', 'ROIC', 'ROE'),
                    #shared_xaxes= True
                    )

fig.add_trace(go.Bar(name='Margem_Ebtida', x=varejo.index, y=varejo['mrgebit']), row=1, col=1)
fig.add_trace(go.Bar(name='Margem_Liquida', x=varejo.index, y=varejo['mrgliq']), row=1, col=2)
fig.add_trace(go.Bar(name='ROIC', x=varejo.index, y=varejo['roic']), row=2, col=1)
fig.add_trace(go.Bar(name='ROE', x=varejo.index, y=varejo['roe']), row=2, col=2)

fig.update_layout(title_text='<b> Overview Rentabilidade Varejo<b>', template='plotly_dark', showlegend=False)
fig.show()
```

```python
varejo2 = df.query('index in @setor').loc[:,['pl', 'pvp', 'psr', 'dy']]
varejo2
```

```python
fig = make_subplots(rows=2, 
                    cols = 2, 
                    row_heights=[5,5],
                    #column_widths=[3,3], 
                    subplot_titles=('Preço/Lucro', 'P/VP', 'PSR', 'Dividend Yeld'),
                    #shared_xaxes= True
                    )

fig.add_trace(go.Bar(name='Preço/Lucro', x=varejo2.index, y=varejo2['pl']), row=1, col=1)
fig.add_trace(go.Bar(name='P/VP', x=varejo2.index, y=varejo2['pvp']), row=1, col=2)
fig.add_trace(go.Bar(name='PSR', x=varejo2.index, y=varejo2['psr']), row=2, col=1)
fig.add_trace(go.Bar(name='Dividend Yeld', x=varejo2.index, y=varejo2['dy']), row=2, col=2)

fig.update_layout(title_text='<b> Overview Rentabilidade Varejo<b>', template='plotly_dark', showlegend=False)
fig.show()
```
